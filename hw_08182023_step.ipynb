```python
#this is for finding the minimum for f=x^2
import matplotlib.pyplot as plt
import numpy as np

x_array,f_array,fmin_array=[],[],[] #original
alpha=0.001; slope=1 #step size #(original slope,how many steps)
x0=1;f0=x0**2;fmin=f0
x_array.append(x0)
f_array.append(f0)
fmin_array.append(fmin)
n=2000 #iteration num
#update
x=x0+alpha
f=x**2
if f<f0:
    fmin=f
x_array.append(x)
f_array.append(f)
fmin_array.append(fmin)

for i in range(1,n,1):
    #use slope
    vp=(f_array[-1]-f_array[-2])/(x_array[-1]-x_array[-2])
    x=x-vp*alpha #pay attemtion to this minum sign
    f=x**2
    if f<fmin:
        fmin=f
    x_array.append(x)
    f_array.append(f)
    fmin_array.append(fmin)

plt.figure(figsize=(5,3))
plt.plot(np.linspace(1,n+1,n+1), fmin_array,"r",label="fmin")
plt.plot(np.linspace(1,n+1,n+1), f_array,label="f")
plt.title("optimization,alpha="+str(alpha)+",iteration="+str(n))
plt.ylabel("",fontsize=10)
plt.xlabel("iteration number",fontsize=10)
plt.tick_params(axis='both', which='major', labelsize=10)
plt.legend()
```
